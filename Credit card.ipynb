```python
df=pd.read_csv("creditcard.csv")
```

```python
df.drop(["Time","V26","V25","V23","V22","V15","V13","V8","V19","V20","V21","V24","V27","V28"],axis=1,inplace=True)
```

```python
df=df.sample(frac=1).reset_index(drop=True)
df1=df[df["Class"]==0][:10000]
df1=df1.append(df[df["Class"]==1][:500])
df1=df1.reset_index(drop=True)
display(df1["Class"].value_counts())
df1
```

```python
sns.countplot(df["Class"])
```

```python
df1.head()
```

# SPLITING DATA.

```python
X=df1.iloc[:,:-1]
y=df1.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=93)
```

# Decision Tree Model Building.

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
y_train_pred=dt.predict(X_train)
y_test_pred=dt.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print("Train Data")
print(classification_report(y_train,y_train_pred))
print("Test Data")
print(classification_report(y_test,y_test_pred))
```

```python
param_grid={
    
     "criterion":["gini"],
     "max_depth":np.arange(1,10),
     "min_samples_split":np.arange(2,50,2),
     "min_samples_leaf":np.arange(2,50),
     "class_weight":[None ,"balanced"]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
random_dt=RandomizedSearchCV(dt,param_distributions=param_grid,scoring="f1",n_jobs=-1)
random_dt.fit(X_train,y_train)
```

```python
random_dt.best_estimator_
```

```python
random_dt.best_params_
```

```python
random_dt.best_score_
```

```python
grid_train_pred1=random_dt.predict(X_train)
grid_test_pred1=random_dt.predict(X_test)
```

```python
print("Train data")
print(classification_report(y_train,grid_train_pred1))
print("Test data")
print(classification_report(y_test,grid_test_pred1))
```

# Logistic regression model

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(class_weight='balanced')
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred_train=lr.predict(X_train)
y_pred_test=lr.predict(X_test)
```

```python
print("Train data")
print(classification_report(y_train,y_pred_train))
print("Test data")
print(classification_report(y_test,y_pred_test))
```
