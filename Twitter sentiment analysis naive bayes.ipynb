```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
```

```python
df=pd.read_csv("tweets.csv.zip")
```

```python
df
```

```python
df.drop(["id","keyword","location"],axis=1,inplace=True)
```

```python
df.head(5)
```

# CLEANING OF TEXT COLUMN.

```python
df["text"]=df["text"].apply(lambda x:" ".join(x.lower() for x in x.split(" ")))
```

```python
from bs4 import BeautifulSoup
```

```python
df["text"]=df["text"].apply(lambda x:BeautifulSoup(x).get_text())
```

```python
import re
df["text"]=df['text'].apply(lambda x:re.sub("https\S+","",x))
```

```python
pip install nltk
```

```python
import nltk
```

```python
df.sample(9)
```

```python
from nltk.corpus import stopwords
```

```python
stopwords=stopwords.words("english")
print(stopwords)
```

```python
df["text"]=df["text"].apply(lambda x:" ".join([x for x in x.split(" ") if x not in stopwords]))
```

# TO REMOVE NON APLHA CHARACTER.

```python
df["text"]=df["text"].apply(lambda x:" ".join([re.sub("[^A-Za-z]+","",x) for x in nltk.word_tokenize(x)]))
```

```python
df["text"]=df['text'].apply(lambda x:" ".join([x for x in x.split(" ") if len(x)>2]))
```

```python
df.sample(10)
```

```python
df.head(5)
```

```python
from nltk.stem import WordNetLemmatizer
lematizer=WordNetLemmatizer()
def lemmatizer_words(text):
    return " ".join([lematizer.lemmatize(word) for word in text.split()])
```

```python
df['text']=df['text'].apply(lambda text: lemmatizer_words(text))
```

```python
df.head(5)
```

```python
df["target"].value_counts()
```

# SPLITTING DATA

```python
X=df[["text"]]
y=df["target"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=18)
```

```python
print("Train Data",X_train.shape,y_train.shape)
```

```python
print("Test Data",X_test.shape,y_test.shape)
```

# COUNTVECTORIZER

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vect=CountVectorizer()
```

```python
X_train_vect=vect.fit_transform(X_train.values.ravel())
```

```python
X_train_vect
```

```python
X_test_vect=vect.transform(X_test.values.ravel())
```

```python
X_test_vect
```

```python
vect.get_feature_names()
```

# LOGISTIC REGRESSION.

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg=LogisticRegression()
```

```python
reg.fit(X_train_vect,y_train)
```

```python
y_train_pred=reg.predict(X_train_vect)
```

```python
y_test_pred=reg.predict(X_test_vect)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print("Train Data")
print(classification_report(y_train,y_train_pred))
print("Test Data")
print(classification_report(y_test,y_test_pred))
```

```python
confusion_matrix(y_train,y_train_pred)
```

```python
confusion_matrix(y_test,y_test_pred)
```

# NAIVE BAYES.

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
multi_nb=MultinomialNB()
```

```python
multi_nb.fit(X_train_vect,y_train)
```

```python
y_train_predict=multi_nb.predict(X_train_vect)
y_test_predict=multi_nb.predict(X_test_vect)
```

```python
print("Train Data")
print(classification_report(y_train,y_train_predict))
print("Test Data")
print(classification_report(y_test,y_test_predict))
```

```python
confusion_matrix(y_train,y_train_predict)
```
