# USING MIN MAX SCALAR TO NORMALISE(bcoz they are skewed) SOME OF YOUR COLUMNS :- i.e Bringing values between 0 to 1

```python
cols=["CRIM","ZN",'AGE','TAX',"B"]
for col in cols:
    minimum=min(B1[col])
    maximum=max(B1[col])
    B1[col]=(B1[col]-minimum)/(maximum-minimum)
```

```python
# FROM HEATMAP WE CAN SEE THAT STRONG CORELATION BETWEEN TAX AND RAD WHICH WILL AFFECT MY MODEL SO I WILL IGNORE RAD
```

```python
X=B1.drop(columns=['Target','RAD'])
```

```python
Y=B1.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
le=LinearRegression()
```

```python
le.fit(X_train,Y_train)
```

```python
from sklearn.model_selection import cross_val_score
le_cv=cross_val_score(le,X,Y,cv=5)
le_cv
```

```python
Y_pred_train=le.predict(X_train)
Y_pred_test=le.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print("Train Data")
print("RMSE:",round(np.sqrt(mean_squared_error(Y_train,Y_pred_train)),2))
print("R2 Score:",round(r2_score(Y_train,Y_pred_train),2))
print("Test Data")
print("RMSE:",round(np.sqrt(mean_squared_error(Y_test,Y_pred_test)),2))
print("R2 Score:",round(r2_score(Y_test,Y_pred_test),2))
```
